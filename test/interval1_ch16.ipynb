```python
# check the r square versus bin size
import matplotlib.pyplot as plt
import matplotlib.pyplot as plt
import matplotlib as mpl
#import matplotlib as mpl
from sklearn.metrics import mean_squared_error,r2_score
from mpl_toolkits.axes_grid1.axes_divider import make_axes_locatable

from scipy.stats import binned_statistic as bstat
def L_error_2d(error_train,L_train,error_valid,L_valid,error_test,L_test, plotname):
    
    
    fig, axs = plt.subplots(2, 2, figsize=(12, 10), sharex='col', sharey='row',
                            gridspec_kw={'hspace': 0.1, 'wspace': 0.1})

    # Get error all
    error_all = np.concatenate((error_train,error_valid,error_test))
    L_all = np.concatenate((L_train,L_valid,L_test))

    xmin = 2.6
    xmax = 6.6
    ymin = -1.5
    ymax = 1.5
    bins = np.linspace(xmin,xmax,num = 101)
    bin_error = np.linspace(xmin,xmax,num = 15)
    
    # Calculate mean in each L bin
    

    for row in range(2):
        for col in range(2):
            ax = axs[row, col]
            if col == 0:
                err = error_all if row == 0 else error_valid
                ll = L_all if row == 0 else L_valid
                plot_name = 'All' if row == 0 else 'Validation'


            else:
                err = error_train if row == 0 else error_test
                ll = L_train if row == 0 else L_test

                plot_name = 'Train' if row == 0 else 'Test'

            pcm = ax.hist2d(ll, err, range =  [[xmin, xmax], [ymin, ymax]],bins = [100,100] ,cmap=plt.cm.viridis)
            def rmse(x):
                return(np.sqrt(np.average(x**2)))

            
            
            stat_rmse = bstat(ll, err, statistic = rmse, bins = bin_error)
            rmsevalue = stat_rmse[0] 
            stat_mean = bstat(ll, err, statistic = 'mean', bins = bin_error)
            mean = stat_mean[0]

            ax.set_xlim(2.6, 6.6)
            ax.set_ylim(ymin,ymax)
            
            
            x_err  = 0.5*(bin_error[1:] + bin_error[:-1])
            ax.errorbar(x_err,mean,rmsevalue,ecolor='r',color='black',elinewidth=1.3,capsize=4,fmt='+')
            
            print(pcm[0].shape)
            
            #print(np.average(pcm[0][:,:],axis = 0))


            # Plot name
            #ax.text(2, 6, plot_name, color='red', fontsize=15)
            ax.set_title(plot_name)


        #         ax_divider = make_axes_locatable(ax)
        #         cax1 = ax_divider.append_axes("right", size="5%", pad="5%")
        
        cb = fig.colorbar(pcm[3], ax=axs[row, :], pad=0.03, shrink=0.75)
        cb.ax.set_title('Obs#')
    xlabel = 'L'
    ylabel = 'Error'
    for ax in axs.flat:
        ax.set(xlabel=xlabel, ylabel=ylabel)
    for ax in axs.flat:
        ax.label_outer()


    fig_name = 'Figures/' + plotname + '.jpg'
    fig_name_eps ='Figures/' + plotname + '.eps'
    fig_name_pdf = 'Figures/' + plotname + '.pdf'
    plt.savefig(fig_name, format='jpg')
    plt.savefig(fig_name_eps, format='eps')
    plt.savefig(fig_name_pdf, format='pdf')

    plt.show()
    
def R2_Lbin(model,x_input,y_real,L, bins):
    
    L_group_index = np.digitize(L,bins)-1
    scores = np.zeros(len(bins)-1)
    for i in range(len(bins) -1):
        print(bins[i])
        score = r2_score(y_real[L_group_index==i],model.predict(x_input[L_group_index==i]).squeeze())
        print(score)
        scores[i] = score
    return scores
    

L_all = np.concatenate((L_train,L_valid,L_test))

a = plt.hist(L_all,bins = 32)
bins = a[1]
 
    
#eflux_model._x_train,eflux_model._x_valid,eflux_model._x_test,
#                   eflux_model._y_train,eflux_model._y_valid,eflux_model._y_test
model = eflux_model._model
error_train = np.array(model.predict(eflux_model._x_train)).squeeze() - eflux_model._y_train

error_valid = np.array(model.predict(eflux_model._x_valid)).squeeze() - eflux_model._y_valid    
    
error_test = np.array(model.predict(eflux_model._x_test)).squeeze() - eflux_model._y_test 
print('train')
scores_train = R2_Lbin(model,eflux_model._x_train,eflux_model._y_train,L_train,bins)
print('valid')
scores_valid = R2_Lbin(model,eflux_model._x_valid,eflux_model._y_valid,L_valid,bins)
print('test')
scores_test = R2_Lbin(model,eflux_model._x_test,eflux_model._y_test,L_test,bins)
plotname = 'hit2d_909keV_v2'

L_error_2d(error_train,L_train,error_valid,L_valid,error_test,L_test, plotname)

# plot the r2 

    
```
